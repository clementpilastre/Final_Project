```python
sql_query_3 = "SELECT model, AVG(kilometer) FROM main_table GROUP BY model ORDER BY AVG(kilometer) ASC;"
sql_result_3 = pd.read_sql_query(sql_query_3, engine)
sql_result_3.head(10)
```

```python
sql_query_4 = "SELECT model, AVG(dollar_price) as price, registration_year FROM main_table  WHERE brand = 'mercedes_benz' GROUP BY registration_year ORDER BY price ASC;"
sql_result_4 = pd.read_sql_query(sql_query_4, engine)
sql_result_4.head()
```
